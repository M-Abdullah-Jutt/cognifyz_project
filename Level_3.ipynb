The 'equity market', also known as the stock market, is a platform where publicly traded companies' shares (or stocks) are bought and sold.
Capital Appreciation occurs when the value of your shares increases over time, leading to profits when you sell them.
Dividends: Some companies distribute some of their profits to shareholders as dividends.

Here, in our dataset, most participants prefer Capital Appreciation. 

A 'mutual fund' is a pool of money collected from many investors and invested in a diversified portfolio of securities, such as stocks, bonds, or other assets.
Its benefits include Diversification, which means reducing investment risk by spreading money across multiple assets.
Most participants choose mutual funds because of Better Returns and Fund Diversification.

'Government bonds' are debt securities issued by a government to raise funds. When you buy a government bond, you're essentially lending money to the government.
Government bonds are generally considered less risky than corporate bonds or stocks, as governments are less likely to default on their debt.
Participants choose government bonds because of Assured Returns and Safe Investment.

A 'fixed deposit' is a type of savings account where you deposit a fixed sum of money for a specific period (e.g., 3 months, 1 year, 5 years).
Generally, FDs offer higher interest rates than regular savings accounts.
Participants select Fixed Deposit because this's Risk-Free and provides Fixed Returns
